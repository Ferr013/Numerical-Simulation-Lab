Cambiare il campionamento delle velocità con una distribuzione di Boltzmann

Esercizio 4.1.1 non salvare velocità (non servono per il Verlet) ma le posizioni precedenti

Esercizio 4.1.2 ricalcolare le velocità confrontantdo la temperatura di riferimento e far ripartire più volte il programma con circa 2000 passi per volta (la prima sarà come quella fornita di default)

Ho cambiato il file input per selezionare se rinormalizzare le velocità dalle direzioni passate o fare il calcolo iniziale.
Sarebbe appropriato fare dei grafici in entrambi i modi, e in più campionando la velocità con una distribuzione di boltzmann (facoltativo)

Le unità d misura sono in unità di LJ (da rivedere, cfr 4.4)


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

f = np.loadtxt("es4.0/output_epot.dat", usecols=(0), delimiter=' ', unpack='true')
f1 = np.loadtxt("es4.0/output_ekin.dat", usecols=(0), delimiter=' ', unpack='true')
f2 = np.loadtxt("es4.0/output_etot.dat", usecols=(0), delimiter=' ', unpack='true')
f3 = np.loadtxt("es4.0/output_temp.dat", usecols=(0), delimiter=' ', unpack='true')


fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex=True, sharey=False,
                        tight_layout=False)

axs[0,0].plot(f)
axs[0,1].plot(f1)
axs[1,0].plot(f2)
axs[1,1].plot(f3)
axs[1,1].axhline(y=1.1, xmin=0, xmax=1, color='orange', label='True Temp')

#axs.set_title('Single run - 10000 steps - initial T=1.9')
axs[0,0].set_title('Potential Energy')
axs[0,1].set_title('Kinetic Energy')
axs[1,0].set_title('Total Energy')
axs[1,1].set_title('Temperature')

plt.show()
```

# Argon

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("es4.4/Argon/es4.4.1/ave_epot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x1, f1, error1 = np.loadtxt("es4.4/Argon/es4.4.1/ave_ekin.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x2, f2, error2 = np.loadtxt("es4.4/Argon/es4.4.1/ave_etot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x3, f3, error3 = np.loadtxt("es4.4/Argon/es4.4.1/ave_temp.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x4, f4, error4 = np.loadtxt("es4.4/Argon/es4.4.1/ave_press.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

fig, axs = plt.subplots(5, 1, figsize=(10, 10), sharex=True, sharey=False,
                        tight_layout=False)

T0=120
sigma = 0.34
kb = 8.61673324*0.00001

axs[0].errorbar(x,f*(T0*kb),yerr=error*(T0*kb))
axs[1].errorbar(x1,f1*(T0*kb),yerr=error1*(T0*kb))
axs[2].errorbar(x2,f2*(T0*kb),yerr=error2*(T0*kb))
axs[3].errorbar(x3,f3*T0,yerr=error3*T0)
axs[4].errorbar(x4,f4*(T0*1.38064852)/(sigma**3)*(10000/101325),yerr=error4*(T0*1.38064852)/(sigma**3)*(10000/101325))

axs[0].set_title('Potential Energy [eV]')
axs[1].set_title('Kinetic Energy [eV]')
axs[2].set_title('Total Energy [eV]')
axs[3].set_title('Temperature [K]')
axs[4].set_title('Pressure [atm]')

plt.show()
```

### <span style="color:blue"> Exercise 4.1 </span>

In this first part we introduced the possibility to restart the simulation from a previous configuration.<br>
We rescaled temperatures to match the target temperature by letting the system run for 200 steps 4 times, each time starting from the configuration achieved in the previous run and matching the velocities in order to compensate for the drop in kinetic energy associeted with the thermalization of the system.
All measurements are expressed in Lennard-Jones units.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

f = np.loadtxt("es4.0/output_epot.dat", usecols=(0), delimiter=' ', unpack='true')
f1 = np.loadtxt("es4.0/output_ekin.dat", usecols=(0), delimiter=' ', unpack='true')
f2 = np.loadtxt("es4.0/output_etot.dat", usecols=(0), delimiter=' ', unpack='true')
f3 = np.loadtxt("es4.0/output_temp.dat", usecols=(0), delimiter=' ', unpack='true')


fig, axs = plt.subplots(2, 2, figsize=(13, 13), sharex=True, sharey=False,
                        tight_layout=False)

axs[0,0].plot(f)
axs[0,1].plot(f1)
axs[1,0].plot(f2)
axs[1,1].plot(f3)
axs[1,1].axhline(y=1.1, xmin=0, xmax=1, color='orange', label='True Temp')

#axs.set_title('Single run - 10000 steps - initial T=1.9')
axs[0,0].set_title('Potential Energy')
axs[0,1].set_title('Kinetic Energy')
axs[1,0].set_title('Total Energy')
axs[1,1].set_title('Temperature')

axs[1,0].set_xlabel("Measures")
axs[1,1].set_xlabel("Measures")

plt.show()
```

### <span style="color:blue"> Exercise 4.2 </span>

Added the possibility to calculate average values of the thermodynamical parameters with their associated error.<br>
To run this simulation we have choose to let the system start from a previously thermalized configuration with $T=1.1$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("es4.2/ave_epot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x1, f1, error1 = np.loadtxt("es4.2/ave_ekin.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x2, f2, error2 = np.loadtxt("es4.2/ave_etot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x3, f3, error3 = np.loadtxt("es4.2/ave_temp.dat", usecols=(0,1,2), delimiter=' ', unpack='true')


fig, axs = plt.subplots(2, 2, figsize=(13, 13), sharex=True, sharey=False,
                        tight_layout=False)

axs[0,0].errorbar(x,f,yerr=error)
axs[0,1].errorbar(x1,f1,yerr=error1)
axs[1,0].errorbar(x2,f2,yerr=error2)
axs[1,1].errorbar(x3,f3,yerr=error3)

axs[0,0].set_title('Potential Energy')
axs[0,1].set_title('Kinetic Energy')
axs[1,0].set_title('Total Energy')
axs[1,1].set_title('Temperature')

axs[1,0].set_xlabel('#Blocks')
axs[1,1].set_xlabel('#Blocks')
plt.show()
```

### <span style="color:blue"> Exercise 4.3 </span>

We added here the possibility to evaluate the pressure of the system (the initial conditions were left untouched from the previous exercise)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

f = np.loadtxt("es4.3/output_press.dat", usecols=(0), delimiter=' ', unpack='true')
x2, f2, error2 = np.loadtxt("es4.3/ave_press.dat", usecols=(0,1,2), delimiter=' ', unpack='true')


fig, axs = plt.subplots(2, 1, figsize=(13, 13), sharex=False, sharey=False,
                        tight_layout=False)

axs[0].plot(f)
axs[1].errorbar(x2,f2,yerr=error2)

axs[0].set_title('Instantaneous Pressure')
axs[1].set_title('Average Pressure')

axs[0].set_xlabel("Measures")
axs[1].set_xlabel("#Blocks")

plt.show()
```

### <span style="color:blue"> Exercise 4.4 </span>

In the following section we present the graphs of some thermodynamic properties of Argon and Krypton in solid, liquid and gas phases. 
<br>Measurements are expressed in real physical units; energies are in eV, temperature is in K and pressure is in atm.
<br>Both simulations reproduce well a real system made with those elements.

# Argon

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("es4.4/Argon/es4.4.1/ave_epot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x1, f1, error1 = np.loadtxt("es4.4/Argon/es4.4.1/ave_ekin.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x2, f2, error2 = np.loadtxt("es4.4/Argon/es4.4.1/ave_etot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x3, f3, error3 = np.loadtxt("es4.4/Argon/es4.4.1/ave_temp.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x4, f4, error4 = np.loadtxt("es4.4/Argon/es4.4.1/ave_press.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

fig, axs = plt.subplots(5, 1, figsize=(13, 18), sharex=True, sharey=False,
                        tight_layout=False)

T0=120
sigma = 0.34
kb = 8.61673324*0.00001

axs[0].errorbar(x,f*(T0*kb),yerr=error*(T0*kb))
axs[1].errorbar(x1,f1*(T0*kb),yerr=error1*(T0*kb))
axs[2].errorbar(x2,f2*(T0*kb),yerr=error2*(T0*kb))
axs[3].errorbar(x3,f3*T0,yerr=error3*T0)
axs[4].errorbar(x4,f4*(T0*1.38064852)/(sigma**3)*(10000/101325),yerr=error4*(T0*1.38064852)/(sigma**3)*(10000/101325))

axs[0].set_title('Potential Energy [eV]')
axs[1].set_title('Kinetic Energy [eV]')
axs[2].set_title('Total Energy [eV]')
axs[3].set_title('Temperature [K]')
axs[4].set_title('Pressure [atm]')

plt.suptitle('Argon - Solid', fontsize=25)
axs[4].set_xlabel('#Blocks')

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
e = np.loadtxt("evolution.dat")
x_e = e[:,0]
y_e = e[:,1]

fig, axs = plt.subplots(1, 1, figsize=(15, 10))

plt.plot(x_e,y_e)
plt.title('Salesman path evolution')
plt.xlabel('#Generation')
plt.ylabel('Length')

plt.show()
```

```python
b = np.loadtxt("first.gene")
x_b = b[:,1]
y_b = b[:,2]

fig, axs = plt.subplots(1, 1, figsize=(10, 10))

plt.plot(x_b,y_b)
plt.title('Salesman first path')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

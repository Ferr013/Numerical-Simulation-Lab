```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
e = np.loadtxt("evolution.dat")
x_e = e[:,0]
y_e = e[:,1]

fig, axs = plt.subplots(1, 1, figsize=(15, 10))

plt.plot(x_e,y_e)
plt.title('Salesman path evolution')
plt.xlabel('#Generation')
plt.ylabel('Length')

plt.show()
```

```python
f = np.loadtxt("first.gene")
x_f = f[:,1]
y_f = f[:,2]

b = np.loadtxt("best.gene")
x_b = b[:,1]
y_b = b[:,2]

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_f,y_f)
axs[0].set_title('Salesman first path')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b,y_b)
axs[1].set_title('Salesman best path')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show()
```

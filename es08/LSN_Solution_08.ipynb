```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("Energy.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('#blocks')
plt.ylabel('<U>')
plt.grid(True)
plt.show()
```

```python
mu = 0.79
sigma = 0.59

x = np.arange(-3,3,0.1)
dx = x[1] - x[0]
den = 2*sigma*sigma;
psi = (np.exp(-(x-mu)**2/den) + np.exp(-(x+mu)**2/den))
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
```

```python
plt.plot(x, psi)
plt.xlabel('#blocks')
plt.ylabel('<U>')
plt.grid(True)
plt.show()
```

# 1S

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x1, y1, z1 = np.loadtxt("1s.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x1, y1, z1, c=z1, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("1s_ave.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('#blocks')
plt.ylabel('<r>')
plt.grid(True)
plt.show()
```

In this exercise we were asked to implement the Metropolis algorithm to sample Hydrogen atoms' 1s and 2p eigenstates.<br>
The plots below show the points' distribution in 3D space and the average radius of the eigenstates (in Bohr's radius $a_0$ terms), computed with a Uniform Distribution and a Gaussian Distribution.<br>
In each run the step is chosen such that the accept/reject rate is around 50%

# 1S

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x1, y1, z1 = np.loadtxt("1s.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x1, y1, z1, c=z1, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("1s_ave.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('#blocks')
plt.ylabel('<r>')
plt.grid(True)
plt.rcParams['figure.figsize'] = [13,7]
plt.show()
```

# 2P

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x2, y2, z2 = np.loadtxt("2p.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x2, y2, z2, c=z2, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()

```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("2p_ave.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('#blocks')
plt.ylabel('<r>')
plt.grid(True)
plt.rcParams['figure.figsize'] = [13,7]
plt.show()
```

Introducing a **Gaussian** transition probability the results are consistent within $2\sigma$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("2p_ave_gauss.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
x1, f1, error1 = np.loadtxt("1s_ave_gauss.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

fig, axs = plt.subplots(2, 1, figsize=(13, 13), sharex=True, sharey=False,
                        tight_layout=False)

axs[0].errorbar(x1,f1,yerr=error1)
axs[1].errorbar(x,f,yerr=error)

axs[0].set_title('1s')
axs[1].set_title('2p')

axs[1].set_xlabel('#blocks')
axs[0].set_ylabel('<r>')
axs[1].set_ylabel('<r>')

axs[0].grid(True)
axs[1].grid(True)

plt.show()
```

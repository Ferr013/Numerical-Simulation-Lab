In this exercise we were asked to evaluate the price of European Call and Put options.<br>
We first evaluated those by sampling directly the price at time $T=1$. Then we tried to evaluate said price by following it's evolution over dicretized times.<br>
The price can be modeled after a Geometric Brownian Motion (GMB)
As we expected the values obtained are consistent with both methods and situated at most within $2\sigma$'s from the analytical result given by Black and Scholes.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("direct_call.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, label = "Simulation")

plt.axhline(y=14.975790778311286, xmin=0, xmax=1, color='orange', label='Black-Scholes')
plt.title("Direct Call")
plt.xlabel('Blocks')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.rcParams['figure.figsize'] = [13,7]
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("direct_put.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, label="Simulation")
plt.axhline(y=5.4595325819072364, xmin=0, xmax=1, color='orange',label='Black-Scholes')
plt.title("Direct Put")
plt.xlabel('Blocks')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.rcParams['figure.figsize'] = [13,7]
plt.show()
```

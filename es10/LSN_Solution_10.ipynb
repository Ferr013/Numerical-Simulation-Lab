```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# Serial

```python
e = np.loadtxt("evolution.dat")
x_e = e[:,0]
y_e = e[:,1]


fig, axs = plt.subplots(1, 1, figsize=(15, 5))

plt.plot(x_e,y_e)
plt.title('Salesman path evolution')
plt.xlabel('#T_step')
plt.ylabel('Best path length')

plt.show()
```

```python
f = np.loadtxt("first.gene")
x_f = f[:,1]
y_f = f[:,2]

b = np.loadtxt("best.gene")
x_b = b[:,1]
y_b = b[:,2]

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_f,y_f)
axs[0].set_title('First path')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b,y_b)
axs[1].set_title('Best path')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show()
```

# Parallel

```python
e = np.loadtxt("parallel/evolution.dat")
x_e = e[:,0]
y_e = e[:,1]

br = np.loadtxt("parallel/best_rank_evo.dat")
x_br = br[:,0]
y_br = br[:,1]

fig, axs = plt.subplots(2, 1, figsize=(15, 10))

axs[0].plot(x_e,y_e)
axs[0].set_title('Salesman path evolution')
axs[0].set_xlabel('#T_step')
axs[0].set_ylabel('Best path length')

axs[1].plot(x_br,y_br)
axs[1].set_title('Best cpu node')
axs[1].set_xlabel('#T_step')
axs[1].set_ylabel('MPI Rank')

plt.show()
```

```python
f_par = np.loadtxt("parallel/first.gene")
x_f_par = f_par[:,1]
y_f_par = f_par[:,2]

b_par = np.loadtxt("parallel/best.gene")
x_b_par = b_par[:,1]
y_b_par = b_par[:,2]

fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x_f_par,y_f_par)
axs[0].set_title('First path')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].plot(x_b_par,y_b_par)
axs[1].set_title('Best path')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show()
```

with the same input.dat file, the parallel code gives better results

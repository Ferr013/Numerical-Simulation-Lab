### <span style="color:blue"> Exercise 1.1 </span>

In this first exercise we are asked to test the Pseudo-Random Number generator through the estimate of two different integrals.<br>
Each integral is evaluated multiple times with a (supposed) uniform distribution in the integration domain.
The error is estimated through "block-sampling" (a technique that will be used extensively throughout this course).<br>
As we expected the error decreases as the number of "throws" of our uniform dice increases.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("es1.1/es1_1_mean.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('# Throws')
plt.ylabel('<r>')
plt.grid(True)
plt.rcParams['figure.figsize'] = [13,7]
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("es1.1/es1_1_var.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('# Throws')
plt.ylabel('$\sigma$')
plt.grid(True)
plt.rcParams['figure.figsize'] = [13,7]
plt.show()
```

To test that we actually used a uniform distribution with our PRNG, we implemented the $\chi^2$ test.<br>
As suggested in the exercise text, we expect a $\chi^2 \sim 100$.<br>
From our data we see that each individual $\chi^2_j$ estimate is distributed around our prediction, therefore we conclude that our distribution can be considered uniform.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f = np.loadtxt("es1.1/es1_1_chi.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.scatter(x,f)
plt.xlabel('j')
plt.ylabel('$\chi^2$')

plt.show()
```

## <span style="color:blue"> es1.2</span>

We extended the random generator library to have an exponential distribution and a Cauchy-Lorentz distribution.
We then verify the (generalized) Central Limit Theorem (CLT): in fact we see that the uniform and exponential distribution tend to a Gaussian profile as $N \rightarrow \infty$, while the Cauchy-Lorentz distribution remain a Lorentian (as it is a stable distribution).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f = np.loadtxt("es1.2/std_N_1.dat", usecols=(0,1), delimiter=' ', unpack='true')
x1, f1 = np.loadtxt("es1.2/std_N_2.dat", usecols=(0,1), delimiter=' ', unpack='true')
x2, f2 = np.loadtxt("es1.2/std_N_10.dat", usecols=(0,1), delimiter=' ', unpack='true')
x3, f3 = np.loadtxt("es1.2/std_N_100.dat", usecols=(0,1), delimiter=' ', unpack='true')


fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex=False, sharey=False,
                        tight_layout=False)

axs[0,0].hist(f, bins=50)
axs[0,1].hist(f1, bins=50)
axs[1,0].hist(f2, bins=50)
axs[1,1].hist(f3, bins=50)
axs[0,0].set_title('Standard Dice N=1')
axs[0,1].set_title('Standard Dice N=2')
axs[1,0].set_title('Standard Dice N=10')
axs[1,1].set_title('Standard Dice N=100')


plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f = np.loadtxt("es1.2/exp_N_1.dat", usecols=(0,1), delimiter=' ', unpack='true')
x1, f1 = np.loadtxt("es1.2/exp_N_2.dat", usecols=(0,1), delimiter=' ', unpack='true')
x2, f2 = np.loadtxt("es1.2/exp_N_10.dat", usecols=(0,1), delimiter=' ', unpack='true')
x3, f3 = np.loadtxt("es1.2/exp_N_100.dat", usecols=(0,1), delimiter=' ', unpack='true')


fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex=False, sharey=False,
                        tight_layout=False)

axs[0,0].hist(f, bins=50)
axs[0,1].hist(f1, bins=50)
axs[1,0].hist(f2, bins=50)
axs[1,1].hist(f3, bins=50)
axs[0,0].set_title('Exponential Dice N=1')
axs[0,1].set_title('Exponential Dice N=2')
axs[1,0].set_title('Exponential Dice N=10')
axs[1,1].set_title('Exponential Dice N=100')

plt.show()
```

## <span style="color:blue"> es1.3</span>

We implemented the Buffon's needle experiment.<br>
To avoid using $\pi$ to calculate $\pi$, we could not use a sine distribution.<br>
Using $\theta = arctan(Rannyu()/Rannyu())$ gives problem for $\theta \rightarrow \pi/2$, as the tangent goes to $\infty$. <br>
Therefore we generated points inside a circumference and evaluated the sine as the ratio of the coordinates of each point in the frame of reference of the center of the circumference.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("es1.3/es1_3.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error,label='Calculated $\pi$')
plt.xlabel('# throws')
plt.ylabel(r'$\pi$')
plt.axhline(y=3.1415926535, xmin=0, xmax=1, color='orange', label='True $\pi$')
legend = plt.legend(loc='lower right', shadow=True, fontsize='x-large')
plt.grid(True)
plt.show()
```

## <span style="color:blue"> es1.2</span>

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f = np.loadtxt("es1.2/std_N_1.dat", usecols=(0,1), delimiter=' ', unpack='true')
x1, f1 = np.loadtxt("es1.2/std_N_2.dat", usecols=(0,1), delimiter=' ', unpack='true')
x2, f2 = np.loadtxt("es1.2/std_N_10.dat", usecols=(0,1), delimiter=' ', unpack='true')
x3, f3 = np.loadtxt("es1.2/std_N_100.dat", usecols=(0,1), delimiter=' ', unpack='true')


fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharex=True, sharey=False,
                        tight_layout=False)

axs[0,0].hist(f, bins=50)
axs[0,1].hist(f1, bins=50)
axs[1,0].hist(f2, bins=50)
axs[1,1].hist(f3, bins=50)
axs[0,0].set_title('Standard Dice N=1')
axs[0,1].set_title('Standard Dice N=2')
axs[1,0].set_title('Standard Dice N=10')
axs[1,1].set_title('Standard Dice N=100')


plt.show()
```
